# **Langkah (Step) 2:**

Baca Image dari URLs - Read Image From URLS

**--Bahasa--**

Pada Langkah ini kita akan membaca image dari URLs, menampilkannya menggunakan OpenCV, perhatikan ada perbedaan ketika membuka image dalam RGB dan BGR. Channel Warna default untuk OpenCV adalah BGR.

**--English--**

In this step we will read images from urls, and display them using openCV, please note the difference when reading image in RGB and BGR format. The default input color channels are in BGR format for openCV. RGB?

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)				#read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)	#convert color to RGB
  final_frame = cv.hconcat((image, image_2))	#concatenate image
  cv2_imshow(final_frame)				#show image
  print('\n')
```

# **Langkah (Step) 3:**

Melihat ukuran File Image

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

# **Langkah (Step) 4**
mengakeses pixel dengan memberi garis diagonal menyilang

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertikal ditengah image
for y in range (tinggi):
  image_3[y,int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)

#membuat garis menyilang diagonal pada image
for y in range(tinggi):
    image_4[y, int(y*(lebar/tinggi))] = [255, 255, 255]
    image_4[y, lebar - 1 - int(y*(lebar/tinggi))] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_4))
cv2_imshow(final_frame)
```

3. Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang
tertentu

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_default = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menentukan posisi dan panjang garis horizontal
garis_posisi = tinggi // 2
garis_panjang = 60

# Menggambar garis horizontal putih di bagian tengah gambar dengan panjang tertentu
image_4[garis_posisi, (lebar - garis_panjang) // 2 : (lebar + garis_panjang) // 2] = [255, 255, 255]

# Menampilkan gambar yang telah diubah
final_frame = cv.hconcat((image_default, image_4))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_default = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menentukan posisi kotak dan ukurannya
kotak_posisi_x = 120
kotak_posisi_y = 130
kotak_lebar = 150
kotak_tinggi = 100

# Menggambar kotak putih di posisi yang telah ditentukan
image_5[kotak_posisi_y:kotak_posisi_y + kotak_tinggi, kotak_posisi_x:kotak_posisi_x + kotak_lebar] = [255, 255, 255]

# Menampilkan gambar yang telah diubah
final_frame = cv.hconcat((image_default, image_5))
cv2_imshow(final_frame)
```
